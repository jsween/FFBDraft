```python
import pandas as pd
```

```python
cutoff_year = 2019
raw_path = "data/raw/"
processed_path = "data/processed/"
file_name_prefix = "yearly_"
file_name_postfix = "_year_summary"
file_type = ".csv"
```

Offensive Yearly by Position

```python
off_plyr_year = pd.read_csv(f"{raw_path}{file_name_prefix}player_stats_offense{file_type}")
```

```python
off_plyr_year.head()
```

```python
cols_to_keep = ["position", "season", "season_type", "passing_yards", "receiving_yards", "rushing_yards", "rush_touchdown", "pass_touchdown", "interception", "fumble", "total_tds", "depth_team", "games_missed", "games_played_season"]
```

```python
off_plyr_yr_reduced = off_plyr_year[cols_to_keep]
off_plyr_yr_reduced.head()
```

```python
off_plyr_yr_filtered = off_plyr_yr_reduced[(off_plyr_yr_reduced["season_type"] == "REG") & (off_plyr_yr_reduced["season"] > cutoff_year)]
off_plyr_yr_filtered.shape
```

```python
off_plyr_yr_filtered.head()
```

```python
off_plyr_yr_filtered.describe()
```

```python
off_plyr_yr_filtered.to_csv(f"{processed_path}off_position{file_name_postfix}{file_type}")
```

Defensive Yearly by Position

```python
def_df = pd.read_csv(f"{raw_path}{file_name_prefix}player_stats_defense{file_type}")
def_df.shape
```

```python
def_df.head()
```

```python
def_col_to_keep = ["position", "season", "solo_tackle", "assist_tackle", "tackle_with_assist", "sack", "safety", "interception", "def_touchdown", "fumble_forced", "depth_team", "games_missed", "games_played_season"]
```

```python
def_df_reduced = def_df[def_col_to_keep]
def_df_reduced.shape
```

```python
def_df_filtered = def_df_reduced[(def_df_reduced["season"] > cutoff_year) & (def_df_reduced["position"].notna())]
def_df_filtered.shape
```

```python
def_df_filtered.head()
```

```python
def_df_filtered.to_csv(f"{processed_path}def_position{file_name_postfix}{file_type}")
```
