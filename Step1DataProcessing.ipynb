kicker

```python
kicker_files = Path(raw_path).glob("player_stats_kicking*.csv")
```

```python
dfs = []
for file in kicker_files:
    df = pd.read_csv(file)
    df["source_file"] = file.name
    dfs.append(df)

k_df = pd.concat(dfs, ignore_index=True)
```

```python
k_df.shape
```

```python
k_df.head()
```

```python
k_df_filtered = k_df[(k_df["season"] > cutoff_year) & (k_df["season_type"] == "REG")]
k_df_filtered.shape
```

```python
kicker_season = (
    k_df_filtered
    .groupby(["season", "team", "player_name"])
    .agg({
        "fg_made": "sum",
        "fg_missed": "sum",
        "fg_made_0_19": "sum",
        "fg_made_20_29": "sum",
        "fg_made_30_39": "sum",
        "fg_made_40_49": "sum",
        "fg_made_50_59": "sum",
        "fg_made_60_": "sum",
        "pat_made": "sum",
        "pat_missed": "sum",
        "week": "count"
    })
    .reset_index()
)
```

```python
kicker_season.head()
```

```python
kicker_season.shape
```

```python
kicker_season["fg_pct"] = (
    kicker_season["fg_made"] / (kicker_season["fg_made"] + kicker_season["fg_missed"])
)
```

```python
kicker_season.to_csv(processed_path + "kicking_position_summary.csv")
```
