```python
import pandas as pd
import sys

sys.path.append('..')
```

```python
processed_path = "data/processed/"
file_type = ".csv"
```

Read in summary stats

```python
off_df = pd.read_csv(processed_path + "off_position_year_summary" + file_type)
dp_df = pd.read_csv(processed_path + "def_position_year_summary" + file_type)
k_df = pd.read_csv(processed_path + "kicking_position_summary" + file_type)
team_df = pd.read_csv(processed_path + "team_defense_summary" + file_type)
```

```python
off_cleaned.season_type.value_counts()
```

```python
off_cleaned = off_cleaned.drop(columns=['Unnamed: 0', 'season_type'])
```
