Offensive Yearly by Position

```python
off_plyr_year = pd.read_csv("data/raw/yearly_player_stats_offense.csv")
```

```python
off_plyr_year.head()
```

```python
cols_to_keep = ["player_id", "position", "season", "season_type", "passing_yards", "receiving_yards", "rushing_yards", "rush_touchdown", "pass_touchdown", "interception", "fumble", "total_tds", "depth_team", "games_missed", "games_played_season"]
```

```python
off_plyr_yr_reduced = off_plyr_year[cols_to_keep]
off_plyr_yr_reduced.head()
```

```python
off_plyr_yr_filtered = off_plyr_yr_reduced[(off_plyr_yr_reduced["season_type"] == "REG") & (off_plyr_yr_reduced["season"] > 2019)]
off_plyr_yr_filtered.shape
```

```python
off_plyr_yr_filtered.head()
```

```python
off_plyr_yr_filtered.to_csv("data/processed/off_position_year_summary.csv")
```
